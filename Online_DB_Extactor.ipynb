```python
import requests
import pandas as pd
import os

# API endpoint
url = 'https://data.ca.gov/api/3/action/datastore_search_sql'

# SQL query
sql_query = 'select * from "26d69565-9d40-4776-b89a-a0ead3f1bf11"'

# Make the API request
response = requests.get(url, params={'sql': sql_query})

# Check if the request was successful
if response.status_code == 200:
    # Parse the JSON response
    data = response.json()
    
    # Extract the results
    results = data['result']['records']
    
    # Convert to DataFrame
    df = pd.DataFrame(results)
    
    # Convert Year to integer
    df['Year'] = df['Year'].astype(int)
    
    # Drop duplicates based on NDC, keeping the row with the highest Year
    df = df.sort_values('Year', ascending=False).drop_duplicates(subset=['NDC Number'], keep='first')
    
    # Group by Year and create separate DataFrames
    grouped = df.groupby('Year')
    
    # Create a 'resources' folder if it doesn't exist
    resources_folder = 'resources'
    os.makedirs(resources_folder, exist_ok=True)
    
    # Save each year's data to a separate CSV file in the resources folder
    for year, group in grouped:
        filename = os.path.join(resources_folder, f'clean_dm_wac_data_{year}.csv')
        group.to_csv(filename, index=False)
        print(f"Created file: {filename}")
else:
    print(f"Error: {response.status_code}")
    print(response.text)
```

```python
import requests
import pandas as pd
import os

# API endpoint
url = 'https://data.ca.gov/api/3/action/datastore_search_sql'

# SQL query
sql_query = 'select * from "26d69565-9d40-4776-b89a-a0ead3f1bf11"'

# Make the API request
response = requests.get(url, params={'sql': sql_query})

# Check if the request was successful
if response.status_code == 200:
    # Parse the JSON response
    data = response.json()
    
    # Extract the results
    results = data['result']['records']
    
    # Convert to DataFrame
    df = pd.DataFrame(results)
    
    # Convert Year to integer
    df['Year'] = df['Year'].astype(int)
    
    # Add a new column with the first word of Drug Product Description
    df['First Word of Description'] = df['Drug Product Description'].str.split().str[0]

    # Drop duplicates based on NDC, keeping the row with the highest Year
    df = df.sort_values('Year', ascending=False).drop_duplicates(subset=['NDC Number'], keep='first')
    
    # Group by Year and create separate DataFrames
    grouped = df.groupby('Year')
    
    # Create a 'resources' folder if it doesn't exist
    resources_folder = 'resources'
    os.makedirs(resources_folder, exist_ok=True)
    
    # Save each year's data to a separate CSV file in the resources folder
    for year, group in grouped:
        filename = os.path.join(resources_folder, f'clean_dm_wac_data_{year}.csv')
        group.to_csv(filename, index=False)
        print(f"Created file: {filename}")
else:
    print(f"Error: {response.status_code}")
    print(response.text)
```
