```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Sample DataFrame creation (you would typically load your data here)
# df = pd.read_csv('your_data.csv')  # Load your data

# Example structure of df (assuming it's similar to your headers)
# df = pd.DataFrame({
#     'brand_name': ['Ozempic', 'Trulicity', 'Rybelsus', 'Ozempic', 'Trulicity', 'Rybelsus'],
#     'spending_year': [2020, 2020, 2020, 2021, 2021, 2021],
#     'total_spending_per_year': [1000000, 950000, 800000, 1200000, 1000000, 900000],
# })

# Calculate the increase in total_spending_per_year
df['spending_increase'] = df.groupby('brand_name')['total_spending_per_year'].diff()

# Filter for the desired brands
filtered_brands = ['Ozempic', 'Trulicity', 'Rybelsus']
filtered_df = df[df['brand_name'].isin(filtered_brands)]

# Ensure the data is sorted by spending_year for proper plotting
filtered_df = filtered_df.sort_values(by=['brand_name', 'spending_year'])

# Create the line plot
plt.figure(figsize=(12, 6))

for brand in filtered_brands:
    brand_data = filtered_df[filtered_df['brand_name'] == brand]
    plt.plot(brand_data['spending_year'], brand_data['spending_increase'] / 1_000_000, marker='o', label=brand)  # Convert to millions

# Add labels and title
plt.xlabel('Spending Year')
plt.ylabel('Spending Increase (Millions)')
plt.title('Spending Increase Over Years for Ozempic, Trulicity, and Rybelsus')
plt.xticks(sorted(filtered_df['spending_year'].unique()))  # Show all spending years
plt.legend(title='Brand Name')
plt.grid()

# Export the plot as a JPG file to the images folder
output_folder = 'images'
os.makedirs(output_folder, exist_ok=True)  # Create the folder if it doesn't exist
plt.savefig(os.path.join(output_folder, 'spending_increase_plot.jpg'), format='jpg', dpi=300)

# Display the plot
plt.tight_layout()
plt.show()
```
